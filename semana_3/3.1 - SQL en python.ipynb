```python
# cargar datos

import pandas as pd

df=pd.read_csv('../data/shoes.csv')

df.head()
```

```python
# seleccion de datos

df=df[['brand', 'colors', 'sizes', 'prices.amountMax', 'imageURLs', 'ean']]
df.head()
```

```python
df=df.dropna().reset_index(drop=True)

df.head()
```

```python
df.info(memory_usage='deep')
```

```python
df.sizes=df.sizes.apply(lambda x: x.split(','))
df.colors=df.colors.apply(lambda x: x.split(','))

df.head()
```

```python
# un registro por talla y color

df=df.explode('sizes').reset_index(drop=True)
df=df.explode('colors').reset_index(drop=True)

df.head()
```

```python
df.info(memory_usage='deep')
```

```python
codigo=pd.read_csv('../data/upc_corpus.csv')
codigo=codigo.dropna().reset_index(drop=True)

codigo.head()
```

```python
df.ean=codigo.ean.iloc[0:len(df)]

df.head()
```

```python
df.columns=['brand', 'color', 'size', 'price', 'url', 'ean']

df.head()
```

```python
# id

df['id']=[i for i in range(len(df))]

df.head()
```

### Inserta datos

```python
#  insert into tabla (columnas) values (valores);


for i in range(len(df)):
    
    insert_query='insert into {} ({}) values {};'\
                 .format('zapatos', ','.join(df.columns), tuple(df.iloc[i].values))
    
    
    insert_query=f"insert into zapatos ({','.join(df.columns)}) values {tuple(df.iloc[i].values)};"
    
    
    cursor.execute(insert_query)
    
    
db.commit()
```

```python
datos=pd.read_sql('''select track_name, 
                            price, 
                            rating_count_tot,
                            user_rating
                            
                     from ratings
                     
                     where price=0 and user_rating>4;''', motor)
datos.head()
```

```python
datos=pd.read_sql('''select prime_genre,
                            sum(rating_count_tot) as total_rating,
                            avg(user_rating) as avg_rating
                     
                     from ratings
                     where price=0
                     
                     group by prime_genre
                     
                     order by total_rating desc
                     limit 10;''', motor)
datos
```

```python
query='''
        select title, year(max(pubdate)) as 'year',
                count(titleauthor.title_id) as num_titles,
                group_concat(concat(authors.au_fname, ' ', authors.au_lname)) as authors
        
        from titles
        
        left join titleauthor
        on titles.title_id=titleauthor.title_id
        left join authors
        on authors.au_id=titleauthor.au_id
        
        group by titles.title
        order by num_titles desc;

'''
```

```python
df_sql=pd.read_sql(query, motor)

df_sql.head()
```
