````python
df=pd.read_excel('../data/Online Retail.xlsx')

df.head()
````

````python
# crear una columna de nulos

df['Null ID']=[i if i%2 else None for i in range(len(df))]

df.head()
````

````python
# renombrar columnas

df.columns
````

````python
df.columns=[c.lower().replace(' ', '_') for c in df.columns]

df.head()
````

````python
n_cols=[c.lower().replace(' ', '_') for c in df.columns]
````

````python
df=df.rename(columns={k:v for k,v in zip(df.columns, n_cols)})    # k=vieja columna, v=nueva columna

df.head()
````

### Selección dentro del dataframe

````python
df[['quantity', 'unitprice', 'revenue']].head(3)   # seleccion por nombre de columna
````

````python
df[df.quantity==6.].head()    # seleccion segun condicion
````

````python
df[~(df.quantity==6.)].head()  # ~es el NO logico, lo que no es 6
````

````python
df[(df.quantity==6) | (df.country=='EIRE')].head()   # multicondicion, | es el O (or) logico
````
