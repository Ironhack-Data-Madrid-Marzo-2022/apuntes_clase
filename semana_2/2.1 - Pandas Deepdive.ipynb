### Aplicación de funciones

````python
def suma_2(x):
    return x+2
````

````python
df.head()
````

````python
%%time

df.quantity=[suma_2(x) for x in df.quantity]

df.head()
````

````python
%%time

df.quantity=list(map(suma_2, df.quantity))

df.head()
````

````python
%%time

df.quantity=list(map(lambda x: x+2, df.quantity))

df.head()
````

````python
%%time

df.quantity=df.quantity.apply(suma_2)   # apply es el map interno de pandas

df.head()
````

````python
%%time

df.quantity=df.quantity.apply(lambda x: x+2 if x<5 else x+10)   # apply es el map interno de pandas

df.head()
````

````python
%%time

df.quantity=df.quantity + 2   # esto es interno de pandas vectorizado

df.head()
````

````python
%%time

df['nueva']=df.quantity / df.unitprice

df.head()
````

### lookup

````python
seasons=pd.DataFrame(np.random.random((10, 4)), 
                     columns=['winter', 'spring', 'summer', 'autumn'])
seasons
````

````python
look_lst=['summer', 'winter','spring', 
      'winter', 'summer', 'autumn',
      'autumn', 'summer', 'autumn', 'spring']
````

````python
seasons['look']=seasons.lookup(seasons.index, look_lst)
````

````python
seasons.index
````

````python
seasons.head()
````

### get

````python
seasons.get(seasons.spring>0.5)
````
