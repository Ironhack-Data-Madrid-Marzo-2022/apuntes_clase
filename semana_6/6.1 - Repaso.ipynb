```python
def funcion_regex(x):
    
    patron=r'vr\d+'

    raw=re.findall(patron, x)[0]

    patron='[0-9]+'
    
    return float(re.findall(patron, raw)[0])
```

```python
df.chorras.apply(funcion_regex).head()
```

**init**

```python
class Aqui():
    
    def extract(self, x):
        
        return x**2
```

```python
Aqui().extract(8)
```

```python
Aqui(8)
```

```python
class Alli():
    
    def __init__(self, a):
        self.a=a
    
    def extract(self, x):
        
        return x**2
```

```python
Alli().extract(8)
```

```python
Alli(1).extract(8)
```
